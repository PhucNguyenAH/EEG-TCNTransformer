```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
EMB_SIZE = 40
DEPTH = 6
```

```python
df = pd.read_csv("log.csv", index_col=False)
df.sort_values(by=['N', 'aStop'],inplace=True, ignore_index=True)
df
```

## Select fixed and variable bandpass filter

```python
df_filter = df.iloc[25:]
df_filter.reset_index(drop=True, inplace=True)
df_filter.drop(columns=["emb_size", "depth"],inplace=True)
# subjects = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 'avg']
# acc_best = 0
print(df_filter.groupby(['N'])['avg'].max())
print(df_filter.iloc[df_filter.groupby(['N'])['avg'].idxmax()])
```

```python
print(df_filter['avg'].max())
print(df_filter.iloc[df_filter["avg"].idxmax()][['N', 'aStop']])
```

```python
col_names = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 'avg']

mask = (df['emb_size']==EMB_SIZE) & (df['depth']==DEPTH) & (df['N']==0) & (df['aStop']==0)
df_baseline = df.loc[mask]
df_baseline_values = df_baseline.values[0]
df_baseline_values[4:] = df_baseline_values[4:]*100

df_fixed_filter = pd.DataFrame(df_filter.iloc[df_filter["avg"].idxmax()])
df_fixed_filter_values = df_fixed_filter.values.reshape(-1)
df_fixed_filter_values[2:] = df_fixed_filter_values[2:]*100

fixed_dict = {}
fixed_dict["subjects"] = col_names
fixed_dict['baseline'] = df_baseline_values[4:]
fixed_dict['fixed bandpass filter'] = df_fixed_filter_values[2:]
print(fixed_dict)
df_fixed = pd.DataFrame(fixed_dict,columns=['subjects','baseline','fixed bandpass filter'])
print(df_fixed)
ax = df_fixed.plot(x="subjects", y=["baseline", "fixed bandpass filter"],
        kind="bar", figsize=(10, 10))
ax.set_ylabel("Accuracy")
# Display plot
plt.show()
```

```python
subjects = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9']
acc_best = 0
for sub in subjects:
    print(df_filter[sub].max())
    acc_best += df_filter[sub].max()
    print(df_filter.iloc[df_filter[sub].idxmax()][['N', 'aStop']])
print("best average accuracy =", acc_best/9)
```

```python
col_names = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 'avg']
subjects = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9']
mask = (df['emb_size']==EMB_SIZE) & (df['depth']==DEPTH) & (df['N']==0) & (df['aStop']==0)
df_baseline = df.loc[mask]
df_baseline_values = df_baseline.values[0]
df_baseline_values[4:] = df_baseline_values[4:]*100

best_subs = []
for sub in subjects:
    best_subs.append(df_filter[sub].max()*100)
avg_sub = sum(best_subs)/9
best_subs.append(avg_sub)
varibale_dict = {}
varibale_dict["subjects"] = col_names
varibale_dict['baseline'] = df_baseline_values[4:]
varibale_dict['varible bandpass filter'] = best_subs
print(varibale_dict)
df_variable_filter = pd.DataFrame(varibale_dict,columns=['subjects','baseline','varible bandpass filter'])
print(df_variable_filter)
ax = df_variable_filter.plot(x="subjects", y=["baseline", "varible bandpass filter"],
        kind="bar", figsize=(10, 10))
ax.set_ylabel("Accuracy")
# Display plot
plt.show()
```

```python
compare_dict = {}
compare_dict["subjects"] = col_names
compare_dict['baseline'] = df_baseline_values[4:]
compare_dict['fixed bandpass filter'] = df_fixed_filter_values[2:]
compare_dict['varible bandpass filter'] = best_subs
df_compare = pd.DataFrame(compare_dict,columns=['subjects','baseline','fixed bandpass filter','varible bandpass filter'])
ax = df_compare.plot(x="subjects", y=["baseline", 'fixed bandpass filter',"varible bandpass filter"],
        kind="bar", figsize=(10, 10))
ax.set_ylabel("Accuracy")
# Display plot
plt.show()
```
